```python
import pandas as pd
import gc
import timeit
```

```python
df = pd.read_csv('../data/fines.csv')
```

Итерации: во всех следующих подзадачах необходимо выполнить вычисления fines/refund*yearдля каждой строки. Создайте новый столбец с вычисленными данными. 

```python
%%timeit
new_col = []

for i in range(0,len(df)):
    new_col.append(df.iloc[i]['Fines'] / (df.iloc[i]['Refund'] * df.iloc[i]['Year']))

df['Fines / Refund * Year'] = new_col
```

```python
%%timeit
new_col = []

for index, row in df.iterrows():
    new_col.append(row['Fines'] / (row['Refund'] * row['Year']))

df['Fines / Refund * Year'] = new_col
```

```python
%%timeit
df['Fines / Refund * Year'] = df.apply(
    lambda row: row['Fines'] / row['Refund'] * row['Year'], axis = 1
)
```

```python
%%timeit
df['Fines / Refund * Year'] = df['Fines'] / (df['Refund'] * df['Year'])
```

```python
%%timeit
df['Fines / Refund * Year'] = df['Fines'].values / (df['Refund'].values * df['Year'].values)
```
