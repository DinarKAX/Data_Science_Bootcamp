```python
import pandas as pd
```

```python
df = pd.read_csv(
    '../data/feed-views.log',
    skiprows=[2, 3],
    skipfooter=2,
    names=['datetime', 'user'],
    index_col='datetime',
    engine= 'python',
    sep= '\t'
)
#пропускает 2 и 3 строки (считая с 0), пропускает последние 2 строки,
#задаёт имена столбцов,
#делает datetime индексом.

df
```
