```python
auto = auto.dropna(thresh=500, axis=1)
auto.isnull().sum()
```

```python
# Заменяем пропущенные значения в столбце Refund предыдущим значением
auto['Refund'] = auto['Refund'].fillna(method='ffill')
auto.isnull().sum()
```

```python
#Заменим все пропущенные значения в столбце Fines средним значением этого столбца
auto['Fines'].fillna(auto['Fines'].mean(), inplace=True)
auto.isna().sum()
```

```python
# Разделите и проанализируйте марку и модель.

# Используйте метод apply для разделения и извлечения значений в новые столбцы, Makeа также Model.
# Удалить столбец Make_n_Model.
# Сохраните фрейм данных в auto.jsonJSON-файл в указанном ниже формате:

auto
```

```python
auto['Make'] = auto['Make_n_model'].apply(lambda x: x.split()[0])
auto['Model'] = auto['Make_n_model'].apply(lambda x: x.split()[1] if len(x.split()) == 2 else None)
auto.drop('Make_n_model', inplace=True, axis=1)

auto
```

```python
auto.to_json('auto.json', orient='records', indent=2)
```
