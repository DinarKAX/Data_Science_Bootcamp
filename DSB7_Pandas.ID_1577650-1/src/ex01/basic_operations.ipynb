```python
import pandas as pd
```

```python
col_names = ['datetime', 'user']
views = pd.read_csv('../data/feed-views.log', sep='\t', names = col_names)
views
```

```python
# Конвертировать datetimeв datetime64[ns] Dtype.
views['datetime'] = pd.to_datetime(views.datetime)
print(views['datetime'].dtype)
```

```python
# Извлечь год, месяц, день, час, минуту и ​​секунду из значений столбца datetimeв новые столбцы.

# Извлечение каждого компонента в отдельный столбец

views['year'] = views['datetime'].dt.year
views['month'] = views['datetime'].dt.month
views['day'] = views['datetime'].dt.day
views['hour'] = views['datetime'].dt.hour
views['minute'] = views['datetime'].dt.minute
views['second'] = views['datetime'].dt.second

views
```

```python
# 2. Создайте новый столбец daytime.

# Присвойте значение времени суток, если час попадает в определённый интервал. Например, присвойте значение «после полудня», если час больше 11, но меньше или равен 17.
# 0–3:59 = ночь, 4–6:59 = раннее утро, 7–10:59 = утро, 11–16:59 = день, 17–19:59 = ранний вечер, 20–23:59 = вечер.
# Используйте метод cutдля решения этой подзадачи.
# Назначьте столбец userв качестве индекса.
label = ['night', 'early morning', 'morning', 'afternoon', 'early evening', 'evening']

bin = [0, 4, 7, 11, 17, 20, 24]
views['daytime'] = pd.cut(views.hour, bins = bin , labels = label, right=False)
views
```

```python
#Назначьте столбец userв качестве индекса.
views = views.set_index('user')
views
```

```python
#3. Рассчитайте количество элементов в вашем фрейме данных.

# Используйте метод count().
# Рассчитайте количество элементов в каждой категории времени суток, используя данный value_counts()метод.

views.count()
```

```python
#Сортируйте значения в таблице данных по часам,
# минутам и секундам в порядке возрастания одновременно, а не по одному.

sort_e = views.sort_values(by=['hour', 'minute', 'second'])

sort_e
```
